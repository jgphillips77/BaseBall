### Let's look at all recent WS Winners:

```python
# Show only Salary of WS Winning Players.

wswin = salaries[(salaries['WS'] == 1)].set_index('yearID')
wswin.head(5)
```

```python
# Use .describe() on Individual Player Salary for WS Winners:
wswin['salary'].describe()
```

### Here we can see a few fun facts:
* Average Salary is 2.77 million.
* Median Salary is 875,000.
* A-Rod's 33 million in 2009 probably skews the results... Still!


```python
### Let's compare these to those that didn't win World Series:
```

```python
no_win['salary'].describe()
```

### Let's look at Total Salary of each WS Winning Team:

```python
# Groupby and .sum() salary to get TOTAL Salary of each WS Winning Team.

ws_winners = wswin.groupby(['yearID']).sum()
ws_winners
```

```python
ws_winners.columns
```

```python
ws_winners.plot()
```

```python
# Look at Teams Stats for WS Winners since 1985:

wsteam = teams[(teams['WSWin'] == 1) & (teams['yearID'] > 1984)]
wsteam.set_index('yearID')
```

```python
ws_salary = pd.DataFrame(ws_winners)
ws_salary.columns
```

```python
ws_team_salary = wsteam.join(ws_salary, on='yearID').set_index('yearID')

# Since the Salaries are rapidly increasing, I only want to look at the most recent 5 years:
ws_team_salary.tail(5)
```

### Now that the join is done, let's look at the key Stats of these recent Teams:

* R: Runs scored.
* RA: Runs Allowed.
* ERA: Earned Run Average.
* SHO: Shutouts.
* PPF: Pitcher 3 year Park Factor.
* OPS: On base Plus Slugging percentage.
* Attendance, Winning Percentage, and Salary thrown in for fun!

```python
ws_team_salary.tail(5)[['R', 'RA', 'ERA', 'SHO', 'PPF', 'OPS', 'attendance', 'WinP','salary']].describe()
```

```python
# Since the Rockies were bad in 2015, I'll look again at their 2016 info:

rt = teams[(teams['yearID'] == 2016) & (teams['franchID'] == 'COL')]
rt.loc[:, ('R', 'RA', 'ERA','PPF', 'SHO', 'OPS', 'attendance', 'WinP')]

# And we know from above that 2016 Rockies Payroll was $112 million.
```

### How do the Rockies Compare?

* Rockies 845 Runs is definitely competitive vs. 753 mean & 724 median.
* Rockies allowed 860 Runs needs improving to be competitve vs. 623 mean & 641 median.
* Likewise Rockies ERA of 4.91 is weak vs. 3.57 mean & 3.68 median.
* PPF of 117 is another weak pitching stat for the rockies, Winners are 96.2 mean & 95.0 median.
* Only 9 Shutouts vs. 11.4 mean & 12.0 median.
* OPS of 0.793 is solid vs. 0.744 mean & 0.733 median.

### How do the Rockies improve?
