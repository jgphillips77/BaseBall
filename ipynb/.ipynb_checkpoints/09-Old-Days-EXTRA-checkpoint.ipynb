### Hierarchial Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet, fcluster
from scipy.spatial.distance import pdist
import numpy as np
import pandas as pd
import os
```

```python
Xhc = Xs.as_matrix(columns=None)
#Xnr = new_reduced.as_matrix(columns=None)
```

```python
Z = linkage(Xhc, 'ward') # From Xhc
Z
```

```python
pd.DataFrame(Z, columns=["cluster_source_link", "cluster_target_link", "source_target_distance", "new_samples"])
```

```python
c, coph_dists = cophenet(Z, pdist(Xhc))
c
```

```python
def plot_dendogram(df):
    
    # Data prep
    X = df.as_matrix(columns=None)
    Z = linkage(X, 'ward')
    
    # plotting
    plt.title('Dendrogram')
    plt.xlabel('Index Numbers')
    plt.ylabel('Distance')
    dendrogram(
        Z,
        leaf_rotation=90.,  
        leaf_font_size=8.,
    )
    plt.show()
    
    
plot_dendogram(old_teams_num) #old_teams_num
```

```python
max_dist = 44

X = Xs[Xs.columns[1:15]].as_matrix(columns=None)
Z = linkage(X, 'ward')
clusters = fcluster(Z, max_dist, criterion='distance')

Xs['cluster'] = clusters
Xs
#Xs.sort_values("cluster")
# lang_clusters = lang.groupby('cluster')
# lang_clusters.agg([np.mean, np.min, np.max, np.size])
```
