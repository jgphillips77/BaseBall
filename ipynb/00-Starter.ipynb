### Work on Salary.csv to look at 2017 Team Salaries

```python
salary.head(12)  # 2017 Season Team Salary
```

```python
salary.isnull().sum()
```

```python
salary.info()
# Have to change Objects to integers to do maths
```

```python
# First remove $pecial-characters:

salary['TOTAL PAYROLL'] = salary['TOTAL PAYROLL'].str.replace('$', '').str.replace(',','')
salary['PITCHING $'] = salary['PITCHING $'].str.replace('$', '').str.replace(',','')
salary['DH $'] = salary['DH $'].str.replace('$', '').str.replace(',','').str.replace('-', '0')
salary['OUTFIELD $'] = salary['OUTFIELD $'].str.replace('$', '').str.replace(',','')
salary['INFIELD $'] = salary['INFIELD $'].str.replace('$', '').str.replace(',','')
salary['CATCHER $'] = salary['CATCHER $'].str.replace('$', '').str.replace(',','')
```

```python
#  After removing $ and -, now change to numeric.

salary['TOTAL PAYROLL'] = pd.to_numeric(salary['TOTAL PAYROLL'], errors='coerce')
salary['PITCHING $'] = pd.to_numeric(salary['PITCHING $'], errors='coerce')
salary['DH $'] = pd.to_numeric(salary['DH $'], errors='coerce')
salary['OUTFIELD $'] = pd.to_numeric(salary['OUTFIELD $'], errors='coerce')
salary['INFIELD $'] = pd.to_numeric(salary['INFIELD $'], errors='coerce')
salary['CATCHER $'] = pd.to_numeric(salary['CATCHER $'], errors='coerce')
```

```python
#Plot Histogram for 2017 Team Salary 
plt.figure(figsize=(12,12))
salary['TOTAL PAYROLL'].plot.hist(bins=30)

# Make Lines to show cutoff for outliers...
sp = salary['TOTAL PAYROLL']

# Add vertical lines at 0, 1, 2, and 3 standard deviations from the mean.
plt.axvline(sp.mean()+ sp.std(), c='green', label = '1 std from mean', dashes = [5,5])
plt.axvline(sp.mean()- sp.std(), c='green', dashes = [5,5])
plt.axvline(sp.mean()+ 2*sp.std(), c='orange', label = '2 std from mean', dashes = [5,5])
plt.axvline(sp.mean()-2*sp.std(), c='orange', dashes = [5,5])
plt.axvline(sp.mean()+3*sp.std(), c='red', label = '3 std from mean', dashes = [5,5])
plt.axvline(sp.mean()-3*sp.std(), c='red', dashes = [5,5])
plt.axvline(sp.mean(), c='black', label = 'mean')

# Add a legend.
plt.legend()
plt.xlabel('total salary $')
plt.ylabel('count')

plt.show()
```
